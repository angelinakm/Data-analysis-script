```python
df_income = pd.read_csv('CAincome.csv')
```

```python
#get the 2017 rows only
df_bool = df_income['REF_DATE']==2017
df_income = df_income[df_bool]
```

```python
df_income = df_income[df_income['GEO'].isin(['Newfoundland and Labrador','Nova Scotia','Prince Edward Island','New Brunswick','Quebec','Ontario',
                    'Manitoba', 'Saskatchewan','Alberta','British Columbia','Yukon','Nunavut', 'Northwest Territories'])]
```

```python
df_bool2 = df_income['UOM']=='Dollars'
df_income = df_income[df_bool2]
df_income.head()
```

```python
df_bool3 = df_income['Family characteristics'] == 'Median total income, all families'
df_income = df_income[df_bool3]
df_income.head()
```

```python
df_income = df_income[['GEO', 'VALUE']]
```

```python
sns.barplot(y = 'GEO',x='VALUE', data = df_income, palette ='flare')
```
